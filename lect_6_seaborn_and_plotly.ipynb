```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
df.head()
```

```python
df[df['Country/Region']=='China']
```

```python
temp_df=df.groupby('Country/Region').sum()
temp_df
```

```python
#Transpose (Switch rows and columns)
# and then target index location 1 and 2:
temp_df=temp_df.T
temp_df
```

```python
#target index location 0 and 1:
temp_df.iloc[0:2]
```

```python
#target all dates except lat and long
temp_df=temp_df.iloc[2:]
temp_df
```

```python
#reset the index to be normal ints again
# "inplace" modifies the original one in contrast to the default (shows the copy)
temp_df.reset_index(inplace=True)
temp_df
```

```python
#rename the "index" column to "date"
temp_df.rename(columns={'index':'Date'},inplace=True)
temp_df
```

```python
#change dates into python datetime dtataype
temp_df['Date']=pd.to_datetime(temp_df['Date'])
```

```python
# make the date the index
temp_df.set_index('Date',inplace=True)
temp_df
```

```python
temp_df.loc['9/10/21']
```

```python
confirmed_df_og=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
confirmed_df=confirmed_df_og.groupby('Country/Region').sum()
```

```python
death_df=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
death_df=death_df.groupby('Country/Region').sum()
```

```python
confirmed_df_og
```

```python
death_df
```

```python
merged_df=confirmed_df.merge(death_df,on='Country/Region',suffixes=['_Confirmed','_Death'])
merged_df.reset_index(inplace=True)
```

```python
sbs.scatterplot('9/14/21_Confirmed','9/14/21_Death',data=merged_df,hue='Country/Region',legend=None)
```
